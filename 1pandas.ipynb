# Outliers

## Using Z- score (Removing Outliers)

```python
df = pd.read_csv("AB_NYC_2019.csv")
data = df.copy()
df.shape
```

```python
df.head()
```

#### In a normal distribution:

-> About 68% of the data falls within one standard deviation of the mean.

-> About 95% of the data falls within two standard deviations of the mean.

-> About 99.7% of the data falls within three standard deviations of the mean.

```python
plt.subplot(1,2,1)
sns.distplot(df["price"])
plt.subplot(1,2,2)
sns.distplot(df["availability_365"])
plt.tight_layout()
plt.show();
```

```python
mean = df["price"].mean()
std = df["price"].std()
mean, std
# if std is less than mean then the data is spread around the mean and VICE VERSA
```

```python
highest_allowed = mean + 3 * std
lowest_allowed = mean - 3 * std
highest_allowed, lowest_allowed
```

```python
# highest allowed value = 873, lowest allowed value= -567. below and above values are outliers.
# This range, from 873 to -567, is often used as a rule of thumb to identify potential outliers in a dataset that is assumed to be 
# approximately normally distributed. Values outside this range might be considered as unusual or worth investigating further.
```

#### Trimming

```python
df[(df["price"] < lowest_allowed) | (df["price"] > highest_allowed)] 
# | (OR) + used to samjh gye hoge
```

```python
df[(df["price"] > lowest_allowed) & (df["price"] < highest_allowed)] 
# data without outlier 
```

### Using z-score

By using z score we not need highest_allowed and lowest_allowed, etc 

z-score = (X - mean)/std

```python
df["z-score"] = (df["price"] - mean)/std
df
```

```python
# these are the outliers using z-score method
df[(df["z-score"] < -3) | (df["z-score"] > 3)]
# used 3 coz 3 is the highest limit About 99.7% of the data falls within three standard deviations of the mean.
```

```python
# Dataset by removing these outliers
df[(df["z-score"] > -3) | (df["z-score"] < 3)]
```

### Capping

Upper Capping (Right Capping):

This involves setting a maximum threshold beyond which any value in the dataset is replaced by the capped value.
It is used to limit extreme high values in a datas

t.
Lower Capping (Left Cappi
n):

This involves setting a minimum threshold below which any value in the dataset is replaced by the capped value.
It is used to limit extreme low values in a .

It is used when suppose the dataset contain crucial information then it is not the useful way to delete the whole row for 
just one value.dataset

```python
upper_limit = mean + 3 * std # changing the values above this to upper_limit
lower_limit = mean - 3 * std # # changing the values below this to lower_limit
upper_limit, lower_limit
```

```python
df["price"] = np.where(df["price"] > upper_limit, upper_limit, np.where(df["price"] < lower_limit, lower_limit,df["price"]))
```

## Using InterQuartile Range (Removing outlier) 

Suppose the data is divided into 4 equal halves. 
The formula is IQR =  Q3 - Q1.  The data range that contains middle 50% of the entries.

The value which are below  Q1 - 1.5IQR  or  above  Q3 + 1.5IQR  are  outliers. This is called the 1.5 X IQR rule.

```python
df = pd.read_csv("AB_NYC_2019.csv")
data = df.copy()
data.shape
```

```python
sns.distplot(data["price"])
```

```python
sns.boxplot(df["price"]) # The blue region shows the IQR
# The lower line is the min value and the upper line is max value(circles above this line are outliers)
# The blue region lower part in Q1 and upper part is Q3
```

```python
Q1 = data["price"].quantile(0.25)
Q3 = data["price"].quantile(0.75)
Q1, Q3 # Q1 is 25% of the data and Q3 is 75% as discussed above
```

```python
# The formula is this only
IQR = Q3 - Q1  
IQR
```

```python
# Other way to find quantile value
data.describe()
```

```python
min = Q1 - 1.5 * IQR
max = Q1 + 1.5 * IQR
min, max
```

```python
data[(data["price"] < min ) |  (data["price"] > max)]
```

### Trimming

```python
# Data with no outliers
new_data = data[(data["price"] > min ) &  (data["price"] < max)]
```

```python
# Comparison between data and new_data
plt.figure(figsize = (16, 8))

plt.subplot(2,2,1)
sns.distplot(data["price"])

plt.subplot(2,2,2)
sns.boxplot(data["price"])


plt.subplot(2,2,3)
sns.distplot(new_data["price"])

plt.subplot(2,2,4)
sns.boxplot(new_data["price"])

plt.tight_layout()
plt.show();
```

### Capping

```python
data["price"] = np.where(data["price"] < min, min, np.where(data["price"] > max, max, data["price"]))
```
