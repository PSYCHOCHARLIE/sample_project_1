## Anatomy of Matplotlib plot

```python
# 0. import matplotlib and get it ready for plotting in Jupyter

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # 10 inch, 10 inch = width and height
# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Sample Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save& show (you save the whole figure and run)
fig.savefig("simple-plot.png") 
```

## TWO OPTIONS FOR SUBPLOT

```python
# Subplot option - 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows =2 , ncols = 2, figsize = (10, 5))

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option - 2

fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

#Plot to each different index
ax[0,0].plot(x,x/2);  # 0,0 represents the first graph
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### WHICH ONE SHOULD I USE?   (PYPLOT VS MATPLOTLIB    OO   METHOD)

* When plotting something quickly, okay to use the pyplot method
* When plotting something advanced, use the  OO  method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# PYPLOT METHOD
over_50.plot(kind='scatter', x='age', y='chol', c='target'); # c = color
# . Here, 'target' is provided, indicating that the color of the markers will be based on the values in the 'target'
```

```python
# OO METHOD
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c= 'target', ax = ax);  # ax = ax meanign plot the above ax data

# ax.set_xlim([45, 100]);
```

```python
## OO METHOD FROM SCRATCH
fig, ax = plt.subplots(figsize=(10,6))

## Plot the data
scatter = ax.scatter(x=over_50['age'], y = over_50['chol'], c = over_50['target']);

# Customize the plot
ax.set(title = 'Heart Disease and Cholesterol Levels', xlabel='Age', ylabel='Cholesterol');

# Add a legend
ax.legend(*scatter.legend_elements(), title = 'Target');  # *scatter is a variable, legend_el = c = over_50,   

# Add a Horizontal line
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(nrows = 2, ncols= 1, figsize= (10 , 10))

# Add data to ax0 
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'])

# Customize ax0
ax0.set(title = 'Heart Disease and cholesterol level', xlabel = 'Age', ylabel = 'Cholesterol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= 'Target')

# Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle = '--') 

# Add data to ax1 
scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c= over_50['target'])

# Customize ax1
ax1.set(title = 'Heart Disease and max heart rate', xlabel = 'Age', ylabel = 'max heart rate')

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= 'Target')

# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(), linestyle = '--') ;

# Add a title to the figure 
fig.suptitle("Heart disease analysis",fontsize=16, fontweight='bold');
```

##  removing the extra lines from x and y axis

```python
# Customizing the y and x axis limitations


# Subplot of chol, age, thalach
fig, (ax0, ax1)= plt.subplots(nrows = 2, ncols= 1, figsize= (10 , 10))

# Add data to ax0 
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c= over_50['target'], cmap='winter')

# Customize ax0
ax0.set(title = 'Heart Disease and cholesterol level', xlabel = 'Age', ylabel = 'Cholesterol')
# Removing the extra lines from left  or  changing the x axis limits
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= 'Target')

# Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle = '--') 

# Add data to ax1 
scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c= over_50['target'], cmap='winter')

# Customize ax1
ax1.set(title = 'Heart Disease and max heart rate', xlabel = 'Age', ylabel = 'max heart rate')
# Removing the extra lines from left  or  changing the x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title= 'Target')

# Add a mean line
ax1.axhline(y=over_50['thalach'].mean(), linestyle = '--') ;

# Add a title to the figure 
fig.suptitle("Heart disease analysis",fontsize=16, fontweight='bold');
```

```python
### Multiple Plots
x = np.linspace(0, 10, 50)

fig, ax = plt.subplots(2, 2)
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title("sine")

y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title("cos")

y=np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title("tan")

y=np.random.randn(1000)
ax[1,1].hist(y, bins=40)   # ax is used to define where you want to plot the graph here at x = 1 and y =1 means at last
ax[1,1].set_title("random")

plt.tight_layout()  # Adjust the padding between the graphs
plt.show()
```
